```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("E:\diabetes.csv")

print(f"Dataset loaded! {df.shape[0]} patients, {df.shape[1]} columns ✅")
print(df.head())
```

```python
medical_columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in medical_columns:
    zero_count = (df[col] == 0).sum()
    print(f"{col}: {zero_count} zero values found and replaced")
    df[col] = df[col].replace(0, df[col].median())

print("\nData cleaned successfully! ✅")
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Outcome', data=df, palette='Set2')
plt.title("Diabetic vs Healthy Patients")
plt.xticks([0,1], ['Healthy', 'Diabetic'])
plt.ylabel("Number of Patients")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Glucose', hue='Outcome', kde=True, palette='Set1')
plt.title("Glucose Levels: Diabetic vs Healthy")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='BMI', y='Glucose', hue='Outcome', palette='Set1', alpha=0.6)
plt.title("BMI vs Glucose — Who is Diabetic?")
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Which factors are most related to diabetes?")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler

features = ['Glucose', 'BMI', 'Age', 'Insulin',
            'BloodPressure', 'DiabetesPedigreeFunction']
X = df[features]
y = df['Outcome']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)

print(f"Training patients: {len(X_train)}")
print(f"Testing patients: {len(X_test)}")
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)

print("=== LOGISTIC REGRESSION RESULTS ===")
print(classification_report(y_test, lr_predictions,
      target_names=['Healthy', 'Diabetic']))
```

```python
dt_model = DecisionTreeClassifier(max_depth=4, random_state=42)
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)

print("=== DECISION TREE RESULTS ===")
print(classification_report(y_test, dt_predictions,
      target_names=['Healthy', 'Diabetic']))
```

```python
plt.figure(figsize=(6,4))
cm = confusion_matrix(y_test, dt_predictions)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Healthy','Diabetic'],
            yticklabels=['Healthy','Diabetic'])
plt.title("Confusion Matrix — How accurate is our model?")
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.show()
```

```python
importance = pd.Series(dt_model.feature_importances_, index=features)
importance.sort_values().plot(kind='barh', color='teal', figsize=(8,5))
plt.title("What factors drive diabetes risk the most?")
plt.xlabel("Importance Score")
plt.tight_layout()
plt.show()
```
